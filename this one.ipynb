```python
import tobler as tob
from tobler import area_weighted
import geopandas as gpd
```

```python
phl_2010 = gpd.read_file("2010_final.geojson")
```

```python
phl_2020 = gpd.read_file("2020_final.geojson")
```

```python
intensive_var = ['%_white_10','%_black_10', '%_asian_pop_10', '%_hispanic_pop_10' ]
extensive_var = ['j_total_pop_10']
```

```python
areal_int_est = tob.area_weighted.area_interpolate(phl_2010, 
                                     phl_2020, 
                                     extensive_variables = extensive_var, 
                                     intensive_variables= intensive_var,
                                     allocate_total = True)
```

```python
areal_int_est.head(10)
```

```python
areal_int_est.to_file('working_doc.geojson', driver = 'GeoJSON')
```
